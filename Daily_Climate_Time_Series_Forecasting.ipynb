Function for windowing

```
def windowed_dataset(series, window_size, batch_size, shuffle_buffer):
    series = tf.expand_dims(series, axis=-1)
    ds = tf.data.Dataset.from_tensor_slices(series)
    ds = ds.window(window_size + 1, shift=1, drop_remainder=True)
    ds = ds.flat_map(lambda w: w.batch(window_size + 1))
    ds = ds.shuffle(shuffle_buffer)
    ds = ds.map(lambda w: (w[:-1], w[1:]))
    return ds.batch(batch_size).prefetch(1)
```

Put the data into numpy array

```
time_train = []
meantemp_train = []

time_val = []
meantemp_val = []

with open('/content/data/DailyDelhiClimateTrain.csv') as csvfile:
     reader = csv.reader(csvfile, delimiter=',')
     next(reader)
     for row in reader:
         time_train.append((row[0]))
         meantemp_train.append((float(row[1])))

with open('/content/data/DailyDelhiClimateTest.csv') as csvfile:
     reader = csv.reader(csvfile, delimiter=',')
     next(reader)
     for row in reader:
         time_val.append((row[0]))
         meantemp_val.append((float(row[1])))

time_train_array = np.array(time_train)
time_val_array = np.array(time_val)
meantemp_train_array = np.array(meantemp_train)
meantemp_val_array = np.array(meantemp_val)
```

```
# DO NOT CHANGE THIS CODE
window_size=64
batch_size=256
shuffle_buffer_size=1000

train_set = windowed_dataset(meantemp_train_array, window_size=window_size,
                             batch_size=batch_size, shuffle_buffer=shuffle_buffer_size)

valid_set = windowed_dataset(meantemp_val_array, window_size=window_size,
                             batch_size=batch_size, shuffle_buffer=shuffle_buffer_size)
```

```
model = tf.keras.models.Sequential([
    tf.keras.layers.LSTM(64,input_shape=[None, 1], return_sequences=True),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(1)
])
```

```
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)

# YOUR CODE
model.compile(optimizer=optimizer, loss='mean_squared_error', metrics=['mae'])

history = model.fit(train_set, epochs=250, validation_data=valid_set,verbose=1)
```

```
import matplotlib.pyplot as plt

# Ambil riwayat pelatihan
loss = history.history['loss']
mae = history.history['mae']
epochs = range(1, len(loss) + 1)

# Plot loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot MAE
plt.subplot(1, 2, 2)
plt.plot(epochs, mae, 'b', label='Training MAE')
plt.title('Training MAE')
plt.xlabel('Epochs')
plt.ylabel('MAE')
plt.legend()

plt.tight_layout()
plt.show()
```
